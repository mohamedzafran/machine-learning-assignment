```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load dataset (ensure the path is correct)
data = pd.read_csv('diabetes.csv')  # Change to your file name

# Preview the data
print(data.head())

# Print the column names to identify the correct ones
print("Columns in the dataset:", data.columns)

# Specify features and target
# Replace 'Glucose' and 'Outcome' with the actual names you see in the printed output
X = data[['Glucose']]  # Example feature (you can add more features if needed)
y = data['Outcome']    # Target variable (indicates diabetes presence)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions using the test set
y_pred = model.predict(X_test)

# Evaluate the model
print("Mean squared error (MSE):", mean_squared_error(y_test, y_pred))
print("R-squared:", r2_score(y_test, y_pred))

# Visualize the results (plotting the line of best fit)
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted')
plt.xlabel('Glucose')
plt.ylabel('Outcome')
plt.legend()
plt.title('Linear Regression: Actual vs Predicted')
plt.show()
```
